```python
from pickle import FALSE
import random

'''
Programmer: Mayar Yaser Mahmoud Qasarwa |202110541
Date: 14-8-2024

'''
from collections import deque
from queue import Queue
class BoardState:
    def __init__(self, board, knight_position, visited=None):                                    #done
        self.board = board
        self.knight_position = knight_position
        self.visited = set(tuple(knight_position)) if visited is None else visited
        self.letters = {}

        for i in range(len(board)):
            for j in range(len(board[i])):
                if board[i][j] != '':
                    self.letters.setdefault(board[i][j], (i, j))



        # iterte the board to define the letters in the board and the position for each one


    def is_goal_state(self):                                                                      #done
        # Goal state is when all letters are visited ==6
        return len(self.visited) == 6

    '''
    set of operators to reach the goal state = (possible_knight_moves, ReArrange )
    Applying an operator to a state transforms it to another state in the state space
    '''

    def possible_knight_moves(self):                                                              #done
        # all possible knight moves
        row, col = self.knight_position
        moves = [
            (row + 2, col + 1), (row + 2, col - 1),
            (row - 2, col + 1), (row - 2, col - 1),
            (row + 1, col + 2), (row + 1, col - 2),
            (row - 1, col + 2), (row - 1, col - 2)
        ]
        # Filter out moves that are out of bounds or already visited and the space positions
        #print('all Moves',moves)
        positions= [value for value in self.letters.values()]

        #print('positions in board:',positions)
        valid_moves = [(r, c) for r, c in moves if (r, c) in positions ]
        #print('valid moves:',valid_moves)
        return valid_moves

    def possible_knight_moves2(self):                                                              #done
        # all possible knight moves
        row, col = self.knight_position
        moves = [
            (row + 2, col + 1), (row + 2, col - 1),
            (row - 2, col + 1), (row - 2, col - 1),
            (row + 1, col + 2), (row + 1, col - 2),
            (row - 1, col + 2), (row - 1, col - 2)
        ]
        # Filter out moves that are out of bounds or already visited and the space positions

        valid_moves = [(r, c) for r, c in moves if r>=0 and r<8 and c>=0 and c<8]

        return valid_moves

    def move_knight(self, new_position):
        # Create a new board state after moving the knight
        new_board = [row[:] for row in self.board]  # Copy the board
        #visited= self.visited.copy()
        visited= self.visited.copy()
        visited.add( new_position)

        return BoardState(new_board, new_position,visited)





    def rearrange_node(self, old_position, new_position):
        new_board = [row[:] for row in self.board]
        i,j= new_position
        k,m= old_position
        letter= new_board[k][m]
        visited= self.visited
        visited.add( new_position)
        visited.discard(old_position)
        new_board[k][m]=''
        new_board[i][j]=letter
        return BoardState(new_board,new_position,visited)

    def print_board(self):

      """Prints the given board as a matrix, replacing empty strings with '-'."""
      for row in self.board:
        print(' '.join(element if element else '-' for element in row))




def dfsv2(start_state):

    stack = [start_state]
    # print('1*',str( [v for v,w in start_state.letters.items() if w==start_state.knight_position]))
    expanded_states = 0
    # print('dfs',start_state.visited)

    while stack:
        current_state = stack.pop()
        expanded_states += 1
        if current_state.is_goal_state():
            end=str( [v for v,w in start_state.letters.items() if w==current_state.knight_position])
            return  end , expanded_states,current_state.board

        for move in current_state.possible_knight_moves():

           if move not in current_state.visited:
                new_state = BoardState( current_state.board,move,current_state.visited)
                new_state = new_state.move_knight(move)
                #p= str( [v for v,w in new_state.letters.items() if w==move])
                stack.append((new_state))

    return None, expanded_states,[]




def d_dfs(start_state):
  ex_dfs=0
  start_position= (start_state.knight_position )
  current_state= BoardState(start_state.board,start_position,{start_position})
  while( len(current_state.visited) < 6 ):


        moves= current_state.possible_knight_moves()
        moves2= current_state.possible_knight_moves2()

        for v in current_state.visited:
          if v in moves2:
            moves2.remove(v)
        modyfied=[]
      #  بفحص شو الأحرف اللي ضايل ما زرتها مشان اعيد ترتيبهن
        for n_v in current_state.letters.values():
          if n_v not in current_state.visited:
            modyfied.append(n_v)

          if n_v in moves2:
            moves2.remove(n_v)
        if moves2 and modyfied:
            #modyfied and moves2:
            m= random.choice(moves2)

            current_state =  current_state.rearrange_node(modyfied[0],m)
            print('move',moves2[0], 'to',modyfied[0])
            # current_state.print_board()
            # print()
  ex_dfs += abs(modyfied[0][0]-m[0])+ abs(modyfied[0][1]-m[1])
  print('expaned foe d_dfs=',ex_dfs)
  return dfsv2(BoardState(current_state.board,start_position,{start_position}))






from collections import deque

def bfs(start_state):
    queue = deque([start_state])
    expanded_states = 0


    while queue:
        current_state = queue.popleft()
        expanded_states += 1
        if current_state.is_goal_state():
            end = str([v for v, w in current_state.letters.items() if w == current_state.knight_position])
            return end, expanded_states, current_state.board

        for move in current_state.possible_knight_moves():
            if move not in current_state.visited:
                new_state = BoardState(current_state.board, move, current_state.visited)
                new_state = new_state.move_knight(move)
                queue.append(new_state)

    return None, expanded_states, []
ex_bfs =0

def b_bfs(start_state):
  ex_bfs=0
  start_position= (start_state.knight_position )
  current_state= BoardState(start_state.board,start_position,{start_position})
  while( len(current_state.visited) < 6 ):


        moves= current_state.possible_knight_moves()
        moves2= current_state.possible_knight_moves2()

        for v in current_state.visited:
          if v in moves2:
            moves2.remove(v)
        modyfied=[]
      #  بفحص شو الأحرف اللي ضايل ما زرتها مشان اعيد ترتيبهن
        for n_v in current_state.letters.values():
          if n_v not in current_state.visited:
            modyfied.append(n_v)

          if n_v in moves2:
            moves2.remove(n_v)
        if moves2 and modyfied:
            #modyfied and moves2:
            m= random.choice(moves2)

            current_state =  current_state.rearrange_node(modyfied[0],m)
            ex_bfs += abs(modyfied[0][0]-m[0])+ abs(modyfied[0][1]-m[1])


            print('move',moves2[0], 'to',modyfied[0])
            # current_state.print_board()
            # print()
  print('expanded for b_bfs=',ex_bfs)
  return bfs(BoardState(current_state.board,start_position,{start_position}))

 #My name is Mayar Qasarwa the  first 6 uniqe letters in the name is: {MAY QSR}
board =[
        ['', 'M', '', '', '', '', '', ''],
        ['', '', '', '', '', '', '', ''],
        ['', '', 'A', '', '', '', '', ''],
        ['', '', '', '', 'R', '', '', ''],
        ['', 'Y', '', '', '', '', '', 'S'],
        ['', '', '', 'Q', '', '', '', ''],
        ['', '', '', '', '', '', '', ''],
        ['', '', '', '', '', '', '', '']
]




#starting_letter must Entered by the user
starting_letter = input('enter the starting letter choose from the set: {M,A,Y,Q,S,R}').upper()

#user can enter only an existing letter as a starting point
while(starting_letter not in {'M','A','Y','Q','S','R'}):
        starting_letter = input('enter another valid starting letter!').upper()

#compute the position on the board of the starting_letters
for i in range(len(board)):
    for j in range(len(board[i])):
        if board[i][j]== starting_letter:
            starting_position=tuple((i, j))
            print('starting Position:',starting_position)
            break

#finally let's intiate our object!!
start_state= BoardState(board,starting_position,{starting_position})
start_state.print_board()
print('--------------------------------------------------------------')

#  First constructor equivalent

    # Run DFS
dfs_solution, dfs_expanded,c= dfsv2(start_state)
if(dfs_solution is None):
  dfs_solution, dfs_expanded,c = d_dfs(start_state)

s= 'start point in this goal state: [\''+starting_letter+'\'] or '+ dfs_solution
print("DFS Solution Path:", s)
print("DFS States Expanded:", dfs_expanded)
BoardState(c,starting_position).print_board()
print('--------------------------------------------------------------')

bfs_solution, bfs_expanded, c = bfs(start_state)
if bfs_solution is None:
    # Handle case where BFS doesn't find a solution
    bfs_solution, bfs_expanded ,c = b_bfs(start_state)


s2 = 'start point in this goal state: [\'' + starting_letter + '\'] or ' + bfs_solution
print("BFS Solution Path:", s2)
print("BFS States Expanded:", bfs_expanded )
BoardState(c, starting_position).print_board()
















```
